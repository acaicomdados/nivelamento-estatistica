```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)

df_iris['target'] = iris.target
df_iris['target_names'] = pd.Categorical.from_codes(iris.target, iris.target_names)

df_iris.head()
```

```python
sns.histplot(x='sepal length (cm)',data=df_iris,hue='target_names',kde=True)
```

```python
sns.histplot(x='petal width (cm)',data=df_iris,hue='target_names',kde=True)
```

```python
df_iris.groupby('target_names').mean()
```

```python
df_iris.groupby('target_names').median()
```

```python
sns.boxplot(x='sepal length (cm)',data=df_iris,y='target_names')
```

```python
sns.violinplot(x='sepal length (cm)',data=df_iris,y='target_names')
```

```python

sns.boxplot(x='petal width (cm)',data=df_iris,y='target_names')
```
